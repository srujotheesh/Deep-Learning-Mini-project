```python
import os

# Set the path to the folder containing the files you want to rename
folder_path = 'C:/Users/jothe/OneDrive/Desktop/Project/GLAUCOMA/dataset/train/normal_train'

# List all files in the folder
file_list = os.listdir(folder_path)

# Initialize a counter for generating unique names
counter = 1

# Define the custom string to include in the new names
custom_string = "ntrain_"
custom_stringl = ".png"

# Loop through the files and rename them
for filename in file_list:
    # Construct a new unique name
    new_name = f"{custom_string}_{counter}{custom_stringl}"  # You can customize the naming pattern

    # Build the full file paths for the old and new names
    old_file_path = os.path.join(folder_path, filename)
    new_file_path = os.path.join(folder_path, new_name)

    # Rename the file
    os.rename(old_file_path, new_file_path)

    # Increment the counter
    counter += 1

print("Files renamed successfully.")
```

```python
import os

train_data_directory = "C:/Users/jothe/OneDrive/Desktop/Project/GLAUCOMA/dataset/train"
test_data_directory = "C:/Users/jothe/OneDrive/Desktop/Project/GLAUCOMA/dataset/test"

train_labels = []
test_labels = []

train_image_paths = []  # Declare the list to store training image paths
test_image_paths = []  # Declare the list to store testing image paths

for class_folder in os.listdir(train_data_directory):
    class_label = class_folder  # This should be the label you assigned to the class
    class_path = os.path.join(train_data_directory, class_folder)
    for image_file in os.listdir(class_path):
        image_path = os.path.join(class_path, image_file)
        train_labels.append(class_label)
        train_image_paths.append(image_path)  # Append image paths to the list

for class_folder in os.listdir(test_data_directory):
    class_label = class_folder  # This should be the label you assigned to the class
    class_path = os.path.join(test_data_directory, class_folder)
    for image_file in os.listdir(class_path):
        image_path = os.path.join(class_path, image_file)
        test_labels.append(class_label)
        test_image_paths.append(image_path)  # Append image paths to the list
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator#to scale the train and test run this also
train_gen=ImageDataGenerator(rescale=1./255)
test_gen=ImageDataGenerator(rescale=1./255)
```

```python
train_generator=train_gen.flow_from_directory(#to know how much train and test data consist.
train_d,
target_size=(28,28),
    batch_size=32,
    class_mode='categorical'
)
test_generator=test_gen.flow_from_directory(
test_d,
target_size=(28,28),
    batch_size=32,
    class_mode='categorical'
)
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28,3)))
```

```python
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
history=model.fit(train_generator,epochs=15,validation_data=test_generator)
```

```python
y_prob=model.predict(test_generator)
y_pred=y_prob.argmax(axis=1)
```

```python
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
import matplotlib.pyplot as plt


# Set the path to the directory containing class subdirectories
test_image_path = 'C:/Users/jothe/OneDrive/Desktop/Project/GLAUCOMA/dataset/train/normal/ntrain__3.png'

# List of class labels
class_labels = ['advanced_Glaucoma', 'early_Glaucoma', 'normal']

# Extract the actual class label from the image path
actual_class_label = test_image_path.split('/')[-2]  # Adjust the index based on your directory structure

# Load and preprocess the test image
img = image.load_img(test_image_path, target_size=(28, 28))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0

# Make predictions
predictions = model.predict(img_array)
predicted_class_index = np.argmax(predictions[0])
predicted_class_label = class_labels[predicted_class_index]
confidence_score = predictions[0][predicted_class_index] * 100

# Load the original image
original_image = Image.open(test_image_path)

# Display the original image, predicted class, and confidence score
plt.imshow(original_image)
plt.title(f'Actual: {actual_class_label}\nPredicted: {predicted_class_label}\nConfidence: {confidence_score:.2f}%')
plt.axis('off')
plt.show()

# Verify if predicted class matches actual class
if predicted_class_label == actual_class_label:
    print('Prediction is correct!')
else:
    print('Prediction does not match the actual class.')
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load a pre-trained ResNet50 model (or other suitable model)
base_model = ResNet50(weights='imagenet', include_top=False)

# Add a global average pooling layer and a dense output layer
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(32, activation='relu')(x)
predictions = Dense(3, activation='softmax')(x)  # Three classes

# Create the fine-tuned model
model2 = keras.Model(inputs=base_model.input, outputs=predictions)

# Freeze the layers of the pre-trained model
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Compile the model
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Fine-tune the model on your glaucoma dataset
history2=model2.fit(train_generator, epochs=15, validation_data=test_generator)
```
